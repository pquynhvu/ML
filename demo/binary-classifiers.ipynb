```python
import numpy as np
np.set_printoptions(suppress=True)

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns 
```

**Binary classifier:**

```python
X = np.array([[3,2],[4,-1], [- 1, 3], [1.5, 1.5], [2, 3.5], [-1.5, -1], [1, -0.5], [2, -3], [4.5, -2]]).reshape(9,2)
df = pd.DataFrame(X, columns=['x1', 'x2'])
```

```python
theta = np.array([- 1, 1.5]).reshape(X.shape[1], 1)
theta_null = 3
h = []

for i in range(X.shape[0]):
    z = np.matmul(theta.T, X[i,:].reshape(X.shape[1],1)) + theta_null
    h.append(np.sign(z)[0][0])
    
df = df.assign(h=h)
```

```python
plt.scatter(df[df['h'] == 1]['x1'], df[df['h'] == 1]['x2'], marker='+', s=150, c='b')
plt.scatter(df[df['h'] == -1]['x1'], df[df['h'] ==-1]['x2'], marker='_', s=150, c='r')

plt.plot(np.linspace(-2, 5, num=10), (np.linspace(-2, 5, num=10) - 3)/1.5, 'g:')
plt.text(0.5, -1.9, 'Hyperplane $\\theta^T x + \\theta_0 = 0$')

plt.quiver(theta_null, 0, theta[0], theta[1], color='r', units='xy', scale=2)
plt.text(2, 1.1, '$\\theta = [-1, 1.5]$')

plt.xlabel('$x_{1}$')
plt.ylabel('$x_{2}$')
plt.title('Binary classifier')
plt.grid()
plt.show() 
```
